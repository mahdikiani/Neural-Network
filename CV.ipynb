```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

rgb = lambda x : cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread('/home/mahdi/Downloads/Knuthojdsmossen_EN-US12064544039_1920x1080.jpg')
# cv2.imshow('img', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(rgb(img))
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
